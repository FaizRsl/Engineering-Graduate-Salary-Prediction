### Specialization

Too many values with less than 10 in other

Add all the value count less than 10, into other

```python
specialization = df['Specialization'].value_counts(ascending = False)
specialization
```

```python
groupToOther = specialization[specialization<=10]
groupToOther
```

```python
def removeSpecialization(value):
    if value in groupToOther:
        return 'other'
    else:
        return value

df['Specialization'] = df['Specialization'].apply(removeSpecialization)
df['Specialization'].unique()
```

### AMCAT Scores

As one can notice, there are a number of -1 values in each of the columns 
Lets replace -1 with a NaN value, then, we will replace all NaN Values with the **Mean** of that specific column.

```python
df = df.replace(-1,np.nan)

columnsNan = [column for column in df.columns if df.isna().sum()[column] > 0]

for column in columnsNan:
    df[column] = df[column].fillna(df[column].mean())
    
df.head()
```
